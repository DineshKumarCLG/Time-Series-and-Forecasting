```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load AirPassengers dataset
url = '/content/AirPassengers.csv'
data = pd.read_csv(url, parse_dates=['Month'], index_col='Month')
data.columns = ['Passengers']

# Plot original data
plt.figure(figsize=(10, 4))
plt.plot(data, label='Original Data')
plt.title('Monthly Air Passengers')
plt.xlabel('Year')
plt.ylabel('Passengers')
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Aggregation: Yearly aggregation
yearly_data = data.resample('A').mean()

plt.figure(figsize=(10, 4))
plt.plot(yearly_data, marker='o', linestyle='-', label='Yearly Aggregated')
plt.title('Yearly Aggregated Data')
plt.xlabel('Year')
plt.ylabel('Average Passengers')
plt.legend()
plt.grid(True)
plt.show()
```

```python

# -------------------------------
# Smoothing: Moving Average
# -------------------------------
data['MA_12'] = data['Passengers'].rolling(window=12).mean()

plt.figure(figsize=(10, 4))
plt.plot(data['Passengers'], label='Original')
plt.plot(data['MA_12'], label='12-Month Moving Average', color='red')
plt.title('Trend Estimation using Moving Average')
plt.xlabel('Year')
plt.ylabel('Passengers')
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Trend Elimination (Deseasonalization)
# Subtract moving average (trend)
data['Deseasonalized'] = data['Passengers'] - data['MA_12']

plt.figure(figsize=(10, 4))
plt.plot(data['Deseasonalized'], label='Deseasonalized (Trend Removed)', color='green')
plt.title('Trend Eliminated Time Series')
plt.xlabel('Year')
plt.ylabel('Passengers (Detrended)')
plt.legend()
plt.grid(True)
plt.show()
```
